## Understanding "Print" command

```python
print("Hello World")
print("Hello","World") # print the result with space
print("Hello"+"World") #print with no space
print("Hello, World")
print(12)
print(45+3)
print(45*3)
print("45ab")
#print("333"+555) #error
#print("ABC"+123) #error
#print(45ab)   #error
print("ABC");print(123);print("Hello Colleague!")  #it does not print in the same line
print("ABC", end="");print(123, end="");print("Hello Colleague!")  #print in the same line but with no space --> this represents nothing ("")
print("ABC", end=" ");print(123, end=" ");print("Hello Colleague!") #print in the same line and with space !!!**Note that:  end = " " , it could be anything inside the double quote such as space " " OR other symbols "+", ",", "abc","*" etc.
print("Hello", end=" "); print("World")
print("I am", 28, "years old.")
```
